# 線性模型 ( Linear Model )

```python
import mglearn
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
mglearn.plots.plot_linear_regression_wave()
```

## 線性回歸 ( Linear regression ) ( 最小平方法 ordinary least squares )

```python
from sklearn.linear_model import LinearRegression

X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

```python
# 斜率參數(w)被存在coef_係數中
# 截距參數(b)被存在intercept_係數中
print('lr.coef_: {}'.format(lr.coef_))
print('lr.intercept_: {}'.format(lr.intercept_))
```

```python
print('Training set score: {:.2f}'.format(lr.score(X_train, y_train)))
print('Test set score: {:.2f}'.format(lr.score(X_test, y_test)))
```

```python
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

```python
print('Training set score: {:.2f}'.format(lr.score(X_train, y_train)))
print('Test set score: {:.2f}'.format(lr.score(X_test, y_test)))
```

## 多變數線性回歸 ( Ridge regression )

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print('Training ser score {:.2f}'.format(ridge.score(X_train, y_train)))
print('Test ser score {:.2f}'.format(ridge.score(X_test, y_test)))
```

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print('Training ser score {:.2f}'.format(ridge10.score(X_train, y_train)))
print('Test ser score {:.2f}'.format(ridge10.score(X_test, y_test)))
```

```python
ridge01= Ridge(alpha=0.1).fit(X_train, y_train)
print('Training ser score {:.2f}'.format(ridge01.score(X_train, y_train)))
print('Test ser score {:.2f}'.format(ridge01.score(X_test, y_test)))
```

```python
plt.plot(ridge.coef_, 's', label="Ridge alpha=1")
plt.plot(ridge10.coef_, '^', label="Ridge alpha=10")
plt.plot(ridge01.coef_, 'v', label="Ridge alpha=0.1")

plt.plot(lr.coef_, 'o', label="LinearRegression")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-25, 25)
plt.legend()
```

### Summary
#### 1. Ridge regression較Linear regression有更多的限制條件(正規化)
#### 2. Ridge regression裡，alpha參數決定模型在簡單/訓練集分配多少比重
#### 3. alpha越高，模型越簡單
#### 4. 訓練集數量少的狀況，Ridge regression測試分數通常較好
